- "어떤 단계별로 흘러가도록 알고리즘을 설계해야 하나"를 공부하는 것이 중요!
- 탐욕법(Greddy Algorithm): 알고리즘의 각 단계에서 그 순간에 최적이라고 생각되는 것을 선택
    - 탐욕법으로 최적해를 찾을 수 있는 문제: 현재의 선택이 마지막 해답의 최적성을 해치지 않을 때

- 이 문제를 해결할 때 탐욕법이 적용되는가? -> 다음과 같은 경우가 성립되면 "탐욕법"으로 풀 수 있음
    - 빌려줄 학생들을 "정해진 순서"로 살펴야 하고
    - 이 "정해진 순서"에 따라
    - 우선하여 빌려줄 방향을 정해야 함

```python
# 방법 1
def solution(n, lost, reserve):
    u = [1] * (n + 2)
    for i in reserve:
        u[i] += 1
    for i in lost:
        u[i] -= 1
    for i in range(1, n + 1):
        if u[i - 1] == 0 and u[i] == 2:
            u[i - 1:i + 1] = [1, 1]
        elif u[i] == 2 and u[i + 1] == 0:
            u[i:i + 2] = [1, 1]
    return len([x for x in u[1:-1] if x > 0])
```

```python
# 방법 2
def solution(n, lost, reserve):
    s = set(lost) & set(reserve) # 두 집합의 교집합: 빌릴 필요가 없는 학생들
    l = set(lost) - s # 하나도 없는 학생들
    r = set(reserve) - s # 빌려줄 수 있는 학생들
    for x in sorted(r): # 시간복잡도의 경우, 반복문 k이지만 sorted(정렬)이 있기 때문에 klogk가 됨
        if x - 1 in l:
            l.remove(x - 1)
        elif x + 1 in l:
            l.remove(x + 1)
    return n - len(l)
```

```python
solution(5, [2, 4], [1, 3, 5])
```
