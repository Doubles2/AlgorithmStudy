```python
def solution(s):
    answer = True
    
    if len(s) not in (4, 6):
        answer = False
    else:
        list_n = [str(x) for x in range(0, 10)]
        
        for x in s:
            if x not in list_n:
                answer = False
            
    return answer
```

```python
print(solution('1234'))
print(solution('a234'))
```
