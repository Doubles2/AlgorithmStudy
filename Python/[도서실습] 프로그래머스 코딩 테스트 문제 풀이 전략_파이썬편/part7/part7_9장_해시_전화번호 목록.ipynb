```python
def solution(phone_book):
    s = set(phone_book)
    
    for num in phone_book:        
        prefix = ""
        for ch in num[:-1]:  
            prefix += ch
            if prefix in s:
                return False
    return True
```

```python
print(solution(["119", "97674223", "1195524421"]))
print(solution(["123","456","789"]))
print(solution(["12","123","1235","567","88"]))
```
