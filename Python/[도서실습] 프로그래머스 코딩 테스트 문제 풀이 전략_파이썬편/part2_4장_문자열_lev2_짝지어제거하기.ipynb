```python
def solution(s):
    answer = -1
    # 2 개의 리스트를 두고, 하나씩 비워가자
    list_front = []
    list_back  = [x for x in list(s)]
    
    for x in s:
        if list_front and list_front[-1] == x:
            list_front.pop()
        else:
            list_front.append(x)

    if not list_front:
        answer = 1
    else:
        answer = 0
    return answer
```

```python
# 테스트 case
print(solution('baabaa'))
print(solution('cdcd'))
```
