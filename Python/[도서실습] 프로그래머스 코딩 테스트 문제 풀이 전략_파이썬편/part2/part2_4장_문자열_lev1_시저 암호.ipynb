```python
def solution(s, n):
    list_lower = [x for x in 'abcdefghijklmnopqrstuvwxyz']
    list_upper = [x.upper() for x in list_lower]

    answer = []
    for txt in s:
        if txt == ' ':
            answer.append(txt)
        elif txt in list_lower:
            tmp_idx = list_lower.index(txt) + n
            tmp_idx = tmp_idx if tmp_idx <= len(list_lower) - 1 else tmp_idx - len(list_lower)
            answer.append(list_lower[tmp_idx])
        else:
            tmp_idx = list_upper.index(txt) + n
            tmp_idx = tmp_idx if tmp_idx <= len(list_upper) - 1 else tmp_idx - len(list_upper)
            answer.append(list_upper[tmp_idx])

    return ''.join(answer)
```

```python
print(solution('AB', 1))
print(solution('z', 1))
print(solution('a B z', 4))
```
