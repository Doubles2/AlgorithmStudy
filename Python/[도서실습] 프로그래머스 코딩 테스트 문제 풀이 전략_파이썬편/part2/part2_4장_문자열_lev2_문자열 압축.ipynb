```python
# 첫번째 시도: 6.1
def solution(s):
    l0 = len(s)
    
    l = len(s)
    for idx in range(1, l+1):
        chunks = [s[i:i+idx] for i in range(0, len(s), idx)]

        tmp_s = ''
        prev_x = chunks[0]
        cnt = 1

        for x in chunks[1:]:
            if x == prev_x:
                cnt+=1
            else:
                tmp_s = tmp_s + str(cnt) + prev_x
                prev_x = x
                cnt = 1
        
        tmp_s = tmp_s + str(cnt) + prev_x
        tmp_s = tmp_s.replace('1','')
        
        if len(tmp_s) < l0:
            l0 = len(tmp_s)
    
    return l0
```

```python
# GPT가 가르쳐 준 방법
# 접근(단위 idx로 자른 뒤 run-length처럼 “연속 동일 chunk”를 cnt+chunk로 압축)은 문제 방향 자체는 맞습니다.
# 1) 가장 큰 버그: tmp_s = tmp_s.replace('1','')
# 2) 불필요하게 큰 범위: for idx in range(1, l+1)
# 3) 성능/스타일: 문자열 + 누적은 비효율 + 변수명 개선
def solution(s):
    n = len(s)
    if n == 1:
        return 1

    best = n

    for unit in range(1, n // 2 + 1):
        prev = s[:unit]
        cnt = 1
        parts = []

        for i in range(unit, n, unit):
            cur = s[i:i + unit]
            if cur == prev:
                cnt += 1
            else:
                parts.append(prev if cnt == 1 else f"{cnt}{prev}")
                prev = cur
                cnt = 1

        parts.append(prev if cnt == 1 else f"{cnt}{prev}")

        best = min(best, len("".join(parts)))

    return best
```

```python
print(solution("aabbaccc"))
print(solution("ababcdcdababcdcd"))
print(solution("abcabcdede"))
print(solution("abcabcabcabcdededededede"))
print(solution("xababcdcdababcdcd"))
```

```python
# 내가 짜려했던 방식과 비슷한 유형
def solution(s: str) -> int:
    n = len(s)
    if n == 1:
        return 1

    best = n

    for unit in range(1, n // 2 + 1):
        prev = s[:unit]
        cnt = 1
        comp_len = 0

        for i in range(unit, n, unit):
            cur = s[i:i + unit]
            if cur == prev:
                cnt += 1
            else:
                comp_len += len(prev) + (len(str(cnt)) if cnt > 1 else 0)
                prev = cur
                cnt = 1

        comp_len += len(prev) + (len(str(cnt)) if cnt > 1 else 0)
        best = min(best, comp_len)

    return best
```
