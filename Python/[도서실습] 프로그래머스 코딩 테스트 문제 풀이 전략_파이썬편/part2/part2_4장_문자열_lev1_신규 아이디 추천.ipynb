```python
def solution(new_id):
    import re
    allowed = re.findall(r'[a-z0-9._-]+', new_id.lower())
    answer = ''.join(allowed)
    answer = re.sub(r'\.{2,}', '.', answer)

    if answer.startswith("."):
        answer = answer[1:]
    if answer.endswith("."):
        answer = answer[:-1]

    if answer == '':
        answer = 'a'

    if len(answer) >= 16:
        answer = answer[:15]
        if answer.endswith('.'):
            answer = answer[:-1]

    if len(answer) <= 2:
        while len(answer) < 3:
            answer += answer[-1]

    return answer
```

```python
print(solution("...!@BaT#*..y.abcdefghijklm"))
print(solution("z-+.^."))
print(solution("=.="))
print(solution("123_.def"))
print(solution("abcdefghijklmn.p"))
```

```python
# == gpt가 고쳐준 내 정답 코드
# rstrip(".") : 끝에 있는 점만 제거(6단계에서 딱 맞음)
# ljust(3, new_id[-1]) : 길이가 3 될 때까지 마지막 문자로 채우기(while 제거)
# if not new_id: : == '' 보다 파이썬스럽고 안전

import re

def solution(new_id: str) -> str:
    # 1) 소문자 치환
    new_id = new_id.lower()

    # 2) 허용 문자만 남기기
    new_id = re.sub(r"[^a-z0-9._-]", "", new_id)

    # 3) 마침표 2개 이상 -> 1개
    new_id = re.sub(r"\.{2,}", ".", new_id)

    # 4) 앞/뒤 마침표 제거
    new_id = new_id.strip(".")

    # 5) 빈 문자열이면 "a"
    if not new_id:
        new_id = "a"

    # 6) 길이 15로 자르고, 끝 마침표 제거
    new_id = new_id[:15].rstrip(".")

    # 7) 길이 3 미만이면 마지막 문자 반복
    new_id = new_id.ljust(3, new_id[-1])

    return new_id
```
