```python
def solution(s):
    # print(len(s))
    
    cnt_remove_0 = 0
    idx = 0

    while True:
        idx+=1
        tmp_s = ''

        for x in s:
            if x == '0':
                cnt_remove_0+=1
                continue
            tmp_s+=x
        
        s = format(len(tmp_s), 'b')
        
        if len(s) == 1:
            break

    return [idx, cnt_remove_0]
```

```python
print(solution('110010101001'))
print(solution('01110'))
print(solution('1111111'))
```

```python
# gpt가 고쳐준 파이썬 답게 + pep8답게
def solution(s: str) -> list[int]:
    removed_zeros = 0
    transforms = 0

    while s != "1":
        transforms += 1

        zeros = s.count("0")
        removed_zeros += zeros

        ones_len = len(s) - zeros
        s = format(ones_len, "b")

    return [transforms, removed_zeros]
```
