```python
def solution(s):
    # 바깥 "{{" "}}" 제거 후, "},{" 기준으로 쪼개기
    parts = s[2:-2].split("},{")
    
    # 각 부분을 숫자 리스트로 변환
    sets_ = [list(map(int, p.split(","))) for p in parts]
    
    # 원소 개수(길이)로 정렬
    sets_.sort(key=len)
    
    answer = []
    seen = set()
    
    # 작은 집합부터 새로 등장한 원소를 하나씩 추가
    for subset in sets_:
        for x in subset:
            if x not in seen:
                seen.add(x)
                answer.append(x)
                break  # 이 문제 구조상 매 단계 새 원소는 1개
    
    return answer
```

```python
print(solution("{{2},{2,1},{2,1,3},{2,1,3,4}}"))
print(solution("{{1,2,3},{2,1},{1,2,4,3},{2}}"))
print(solution("{{20,111},{111}}"))
print(solution("{{123}}"))
print(solution("{{4,2,3},{3},{2,3,4,1},{2,3}}"))
```

```python
# gpt 가 알려주는 짧은 방법;
import re
from collections import Counter

def solution(s: str) -> list[int]:
    nums = map(int, re.findall(r"\d+", s))
    return [x for x, _ in Counter(nums).most_common()]
```
