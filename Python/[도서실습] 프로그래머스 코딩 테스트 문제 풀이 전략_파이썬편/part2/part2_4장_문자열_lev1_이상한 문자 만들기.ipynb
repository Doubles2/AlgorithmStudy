```python
def solution(s):
    answer = []
    idx = 0

    for x in s:
        if x == ' ':
            answer.append(' ')
            idx = 0
        else:
            answer.append(
                x.upper() if idx % 2 == 0 else x.lower()
            )
            idx+=1

    return ''.join(answer)
```

```python
print(solution("try hello world"))
print(solution("try hello  world"))
```
