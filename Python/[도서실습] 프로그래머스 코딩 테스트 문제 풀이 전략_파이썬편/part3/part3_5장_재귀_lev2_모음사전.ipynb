```python
def solution(word):
    vowels = ['A', 'E', 'I', 'O', 'U']
    count = 0
    answer = 0

    def dfs(cur):
        nonlocal count, answer

        if cur:
            count += 1
            if cur == word:
                answer = count
                return True

        if len(cur) == 5:
            return False

        for ch in vowels:
            if dfs(cur + ch):
                return True

        return False

    dfs("")
    return answer
```

```python
print(solution('AAAAE'))
print(solution('AAAE'))
print(solution('I'))
print(solution('EIO'))
```
