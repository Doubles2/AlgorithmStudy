```python
# [풀이 핵심]
# 1. 0을 만날 때마다 "나중에 1을 만나면 pair가 될 후보"로 누적한다.
# 2. 1을 만나면, 지금까지 나온 모든 0과 passing pair가 된다.
# 3. 따라서 배열을 한 번만 순회하면서
#    - 0이면 개수 저장
#    - 1이면 저장된 0의 개수만큼 정답에 더한다.
# 4. 누적 pair 수가 1,000,000,000을 초과하면 -1 반환
#
# [핵심 관점]
# "각 0마다 뒤의 1을 세지 말고,
#  각 1마다 앞의 0 개수를 더한다."
#
# [시간복잡도]
# O(N)
```

```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(A):
    east_go = 0
    pair = 0

    for x in A:
        if x == 0:
            east_go+=1
        else:
            pair+=east_go
        
        if pair > 1000000000:
            return -1
    
    return pair
```

```python
# 1. 최소 길이
print(solution([0]))          # expected: 0))
print(solution([1]))          # expected: 0))

# 2. 차가 아예 없는 방향 조합
print(solution([0, 0, 0, 0])) # expected: 0))
print(solution([1, 1, 1, 1])) # expected: 0))

# 3. 가장 단순한 passing 발생
print(solution([0, 1]))       # expected: 1))

# 4. passing이 절대 발생하지 않는 정반대 순서
print(solution([1, 0]))       # expected: 0))

# 5. 작은 예시지만 여러 번 누적되는 경우
print(solution([0, 1, 0, 1, 1])) # expected: 5))

# 6. 모든 0이 앞, 모든 1이 뒤 -> 최대한 많이 발생하는 패턴
print(solution([0, 0, 0, 1, 1, 1])) # expected: 9))

# 7. 모든 1이 앞, 모든 0이 뒤 -> 0이어야 함
print(solution([1, 1, 1, 0, 0, 0])) # expected: 0))

# 8. 번갈아 나오는 패턴
print(solution([0, 1, 0, 1, 0, 1])) # expected: 6))

# 9. 0이 맨 앞에만 있는 경우
print(solution([0, 1, 1, 1, 1]))    # expected: 4))

# 10. 0이 맨 뒤에만 있는 경우
print(solution([1, 1, 1, 1, 0]))    # expected: 0))

# 11. 결과가 정확히 1,000,000,000인 경우
# 0이 25,000개, 1이 40,000개
print(solution([0] * 25000 + [1] * 40000))
# expected: 1000000000

# 12. 결과가 1,000,000,000 초과하는 경우 -> -1
# 0이 50,000개, 1이 50,000개
print(solution([0] * 50000 + [1] * 50000))
# expected: -1
```
