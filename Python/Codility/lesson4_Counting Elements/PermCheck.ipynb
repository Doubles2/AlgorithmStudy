```python
# 시도2 -> 실패 83점
# == GPT 풀이 ==
# permutation(1..N) 조건은 두 가지가 동시에 필요합니다.
# 모든 값이 1..N 범위 안에 있어야 함
# 중복 없이 1..N이 각각 정확히 한 번 나와야 함
# 그런데 합만 보면,
# 어떤 값이 2번 나오고(중복)
# 다른 값이 빠져도(누락)
# 합은 우연히 같게 만들 수 있어요.

def solution(A):
    # Implement your solution here
    total_sum = ( max(A) * (max(A) + 1) / 2)
    list_sum = sum(A)    
    
    answer = total_sum - list_sum
    
    return 1 if answer == 0  else 0
```

```python
def solution(A):
    N = len(A)

    # 1) 범위 체크: 1..N 밖이 있으면 바로 0
    for x in A:
        if x < 1 or x > N:
            return 0

    # 2) 중복 체크: set 크기가 N이면 중복 없음
    return 1 if len(set(A)) == N else 0
```

```python
print(solution([4, 1, 3, 2]))
print(solution([4, 1, 3]))
print(solution([1, 3]))
print(solution([1, 2]))
```
