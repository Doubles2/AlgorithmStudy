```python
# 시도1 -> 실패 25점

def solution(A):
    # Implement your solution here
    total_list = [x for x in range(1, max(A) + 1)]
    
    # 차집합 이용
    answer = [x for x in total_list if x not in A]
    
    return 0 if answer else 1
```
