```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(N, A):
    # Implement your solution here
    answer = [0] * N

    for x in A:
        if x > 5:
            list_max = max(answer)
            answer = [list_max] * N
        else:
            answer[x-1]+=1
    return answer
```

```python
print(solution(5, [3, 4, 4, 6, 1, 4, 4]))
```
