```python
def solution(X, A):
    seen = set()
    remain = X

    for t, pos in enumerate(A):
        if 1 <= pos <= X and pos not in seen:
            seen.add(pos)
            remain -= 1
            if remain == 0:
                return t

    return -1
```

```python
print(solution(6, [1, 3, 1, 4, 2, 3, 5, 4]))
print(solution(5, [1, 3, 1, 4, 2, 3, 5, 4]))
```
