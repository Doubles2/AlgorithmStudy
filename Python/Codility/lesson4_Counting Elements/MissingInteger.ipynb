```python
def solution(A):
    A = [x for x in A if x > 0]
    if not A: return 1

    A.sort()
    prev = A[0]

    if prev != 1:
        return 1

    for x in A[1:]:
        if prev + 1 < x:
            return prev + 1
        else:
            prev = x

    return prev + 1
```

```python
print(solution([1, 3, 6, 4, 1, 2])) # 5
print(solution([1, 2, 3])) # 4
print(solution([-1, -3])) # 1
print(solution([3]))
print(solution([94]))
print(solution([1000000]))
```
