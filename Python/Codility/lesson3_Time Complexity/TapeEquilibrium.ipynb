```python
def solution(A):
    pre_sum = 0
    post_sum = sum(A)
    answer = []

    for x in A[:-1]:         
        pre_sum += x
        post_sum -= x
        answer.append(abs(pre_sum - post_sum))

    return min(answer)
```

```python
print(solution([3, 1, 2, 4, 3]))
print(solution([-1000, 1000]))
print(solution([1000, -1000]))
```
