```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(A):
    # Implement your solution here
    max_v = len(A) + 1
    list_sum = sum(A)
    total_sum = (max_v * (max_v + 1)) // 2
    return total_sum - list_sum
```

```python
print(solution([2, 3, 1, 5]))
```
