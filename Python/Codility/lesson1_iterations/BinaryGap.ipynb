```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(N):
    # Implement your solution here
    answer = 0
    cnt_0 = 0

    s = str(format(N, 'b')) # 이진수 변환

    for x in s:
        
        if x == '1':
            if answer < cnt_0:
                answer = cnt_0
            cnt_0 = 0
        else:
            cnt_0+=1
    return answer
```

```python
print(solution(9))
print(solution(529))
print(solution(20))
print(solution(15))
print(solution(32))
```
