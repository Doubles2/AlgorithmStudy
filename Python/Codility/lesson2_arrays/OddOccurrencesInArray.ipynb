```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(A):
    # Implement your solution here
    from collections import Counter
    counter_a = Counter(A)
    for k, v in counter_a.items():
        if v % 2 == 1:
            return k
```

```python
print(solution([9,3,9,3,9,7,9]))
```
